# **Linear Approximations:**

In this section we’re going to take a look at an application not of derivatives but of the tangent line to a function. Of course, to get the tangent line we do need to take derivatives, so in some way this is an application of derivatives as well.

Given a function, $f(x)$, we can find its tangent at $x=a$. The equation of the tangent line, which we’ll call $L(x)$ for this discussion, is,
$$L(x)=f(a)+f'(a)(x-a)$$

Take a look at the following graph of a function and its tangent line.

graph.jpg

From this graph we can see that near $x=a$ the tangent line and the function have nearly the same graph. On occasion we will use the tangent line, $L(x)$, as an approximation to the function, $f(x)$, near $x=a$. In these cases we call the tangent line the linear approximation to the function at $x=a$.

Let’s take a look at an example.

**Example 1:** Find the linear approximation $f(x)=x^4-6x^3+3x-7$ at $x=-3$.

**Solution:** We will need to take derivative of a function $f(x)$ and some function values.
$$f'(x)=4x^3-18x^2+3, f'(-3)=4(-3)^3-18(-3)^2+3=-267, f(-3)=227$$

There really isn’t much to do at this point other than write down the linear approximation.
$$L(x)=227-267(x+3)=-267x-574$$

Using `sympy` we can verify our results

```python
import sympy as smp
x=smp.symbols('x')
f=x**4-6*x**3+3*x-7
f.subs(x,-3) #Function value at x=-3
```

```python
#Find the derivative at x=-3
df=smp.diff(f,x)
df.subs(x,-3)
```

We can plot the function and linear approximation

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(x):
  return x**4-6*x**3+3*x-7

def g(x):
  return -267*x-574

xlist = np.linspace(-4,5,100)
ylist1 = f(xlist)
ylist2 = g(xlist)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)', color=color)
ax1.plot(xlist, ylist1, color=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('g(x)', color=color)  # we already handled the x-label with ax1
ax2.plot(xlist, ylist2, color=color)


plt.show()
```

**Example 2:** Find the linear approximation to $f(x)=x^{\frac{1}{4}}$ at $x=2$. Use the linear approximation to approximate the value of $11^{\frac{1}{4}}$. Compare the approximated values to the exact values.

**Solution:** Here is the linear approximation.
$$L(x)=1.189207+0.1486(x-2)$$
$$\implies L(x)=0.1486x+0.892007$$

```python
import sympy as smp
x=smp.symbols('x')
f=x**(1/4)
f.subs(x,2) #Function value at x=2
```

```python
smp.diff(f,x).subs(x,2) #Derivative at x=2
```

Finally, here are the approximations of the values along with the exact values.

```python
#Linear approximation function
L=0.1486*x+0.892007
```

```python
#Now find the L(11) and f(11)
LA_value=L.subs(x,11)
LA_value
```

```python
#Function value at x=11
f_value=f.subs(x,11)
f_value
```

```python
#Percentage error
RE=abs(LA_value-f_value)/f_value
Percent_error=RE*100
Percent_error
```

**Example 3:** Determine the linear approximation for $sin(\theta)$ at $\theta=0$.


**Solution:** All that we need to do is compute the tangent line to $sin(\theta)$ at $\theta=0$.
$$f(\theta)=sin(\theta)\rightarrow f'(\theta)=cos(\theta)$$
$$f(0)=0~~~~~f'(0)=1$$
The linear approximation is,
$$L(\theta)=f(0)+f'(0)(\theta-0)$$
$$=0+(1)(\theta-0)$$
$$=\theta$$

So, as long as $\theta$ stays small we can say that $sin(\theta)\approx \theta$.
